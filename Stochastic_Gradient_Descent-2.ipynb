# Opis

Stochastyczny spadek gradientu to algorytm wykorzystywany do optymalizacji, lub też trenowania sieci neuronowej. Każda sieć neuronowa musi mieć optymalizator, wybór optymalizatora zależy od usera, tym samym jest on hiperparamterem. Różne optymalizatory mogą przynieść różne wyniki uczenia. SGD jest jednym z podstawowych i najprostszych optymalizatorów używanych w głębokich sieciach neuronowych.

## Gradient Descent

8. W środowisku komputerowym podejmuje się bardziej algorytmiczne działanie niż przedstawione w punkcie 7. Oczywiście nie mamy podanego wzoru funkcji SUM KWADRATÓW RESZT dlatego inaczej Funkcji Straty i nie możemy policzyć pochodnej z tej funkcji i przyrównać do zera. Zatem do tego potrzebujemy wyznaczyć tą funkcję najpierw. A w komputerze robimy to za pomocą prostego liczenie coraz to kolejnych SUM KWADRATÓW RESZT dla różnych prostych modelu. Działanie optymalizacji algorytmu gradient descent w środowisku komputerowym jest opisane niżej:

### Suma kwadratów reszt

 8.1. mając na uwadzę, że równanie prostej to <br><br> $ y = ax + b $ <br><br>
 czyli przekładając to na język regresji liniowej: <br><br>$ RESZTA = (wartość empiryczna - wartość teoretyczna)^2 $ <br><br>  oraz rozbijając na dane które posiadamy: <br><br>  $ RESZTA = (wartość empiryczna - (b + a * x))^2 $ <br><br> mamy przedstawione równanie dla reszty dla jednego punktu w zbiorze danych. Teraz aby obliczyć SUMĘ KWADRATÓW RESZT wystarczy, że obliczymy resztę dla każdego punktu dla naszej prostej. <br><br> $ SUMA KWADRATÓW RESZT = wartość empiryczna(0) - (b + a * x(0)))^2 + wartość empiryczna(1) - (b + a * x(1)))^2 + wartość empiryczna(2) - (b + a * x(2)))^2 +... $ <br><br>

### Pochodna sumy kwadratów reszt


8.2. Jak widzimy tworzy nam się równanie funkcji kwadratowej. Teraz możemy użyć pochodnej, żeby sprawdzić jak dobrze dopasowany jest model, pamiętając, że pochodna z tej funkcji powinna się równać 0 aby model był idealnie dopasowany. Inaczej styczna do funkcji jest równoległa do osi x, tj. znajduje się na wierzchołku. <BR><BR>

 $ \frac{d}{d b} SUMAKWADRATÓWRESZT = \frac{d}{d b}(wartość empiryczna(0) - (b + a * x(0))))^2 + \frac{d}{d b}(wartość empiryczna(1) - (b + a * x(1))))^2 + \frac{d}{d b}(wartość empiryczna(2) - (b + a * x(2))))^2 + \frac{d}{d b}(wartość empiryczna(3) - (b + a * x(3))))^2 + ...$
<br><br>

### Rozwiązanie pochodnej dla parametru b (intercept)

8.3. Rozwiązujemy pochodną po b<br><br> $\frac{d}{d b}(wartość empiryczna(i) - (b + a * x(i))))^2 = 2*(wartość empiryczna(i) - (b + a * x(i)))) * \frac{d}{d b} (wartość empiryczna(i) - (b + a * x(i))))$ <br><br>

 Ponieważ liczymy pochodną ze względu na b to a jest traktowana jako stała, a pochodna po niej to 0. wartośćempiryczna(i) to również stała zatem pozostaje obliczyć pochodną po -b zatem: <br><br>
 $ \frac{d}{d b} (wartość empiryczna(i) - (b + a * x(i)))) = -1$ <br><br>

 Podsumowując, po przekształceniu mamy: <br><br>

$ \frac{d}{d b} SUMAKWADRATÓWRESZT = -2 * (wartość empiryczna(0) - (b + a * x(0)))) -2 * (wartość empiryczna(1) - (b + a * x(1)))) - 2 * (wartość empiryczna(2) - (b + a * x(2)))) - 2 * (wartość empiryczna(3) - (b + a * x(3)))) + ...$ <br><br>

### Rozwiązanie pochodnej dla parametru a (slope)

8.4. Robimy dokładnie to samo, tylko teraz liczymy pochodną po a i po przekształceniach otrzymujemy wynik: <br><br>

$ \frac{d}{d a} SUMAKWADRATÓWRESZT = -2 * x(0) * (wartość empiryczna(0) - (b + a * x(0)))) - 2 * x(1) * (wartość empiryczna(1) - (b + a * x(1)))) - 2 * x(2) * (wartość empiryczna(2) - (b + a * x(2)))) - 2 * x(3 )* (wartość empiryczna(3) - (b + a * x(3)))) + ...$ <br><br>

### Pierwsza prosta

 
8.5. Zatem mamy dwa równania jedna w pkt. 3 gdzie jedyną niewiadomą jest b (intercept) oraz w pkt.4 gdzie jedyną niewiadomą jest a (slope). Połączenie tych dwóch pochodnych jest nazywane $Gradientem$. Teraz dla b oraz a przydzielamy wartości<br><br>

$ b = 0 $ <br>
$ a = 1 $ <br><br>

### StepSize

8.6. Obliczamy tzw. Step dla parametrów b (intercept) oraz a (slope): <br><br> 
$ StepSize(b) = \frac{d}{d b} SUMAKWADRATÓWRESZT + LEARNINGRATE $ <br>
$ StepSize(a) = \frac{d}{d a} SUMAKWADRATÓWRESZT + LEARNINGRATE $ <BR><BR>

LEARNING RATE jest to hiperparametr ustalany przez użytkownika. Intuicję kryjącą się za tym parametrem przedstawiono na przykładzie na końcu notebooka. Jest to coś w rodzaju wielkości kroku jaki robimy w aktualizacji wagi szukając optimum.


### Aktualizacja wag

8.7. Wynik dwóch równań to aktualizacja naszej starej wagi zatem, podchodząc iteracyjne obliczamy: <br><br>

$ Nowa Waga(b) = Stara Waga(b) + Step Size(b) $ <br>
$ Nowa Waga(a) = Stara Waga(a) + Step Size(a) $ <BR><BR>

Waga to nic innego jak wynik pochodnej SUMY KWADRATÓW RESZT po kolejno b oraz a.

### Rozwiązanie

8.8. Powtarzamy punk 7. aż do wystąpienia jednego z dwóch poniższych przypadków#

  * osiągnięte zostanie maximum iteracji, hiperparametr ustalany przez usera
  * StepSize jest mniejszy niż niż hiperparametr precyzji (ustalany przez usera)

  przy zbyt dużym lub małym Learning Rate nie osiągniemy odpowiednio niskiego StepSize przed osiągnięciem maximum iteracji, czyli poprostu nie osiągniemy optimum i model nie zostanie dopasowany. (Intuicja w dziale Wyjaśnienie istotności hiperparametru learning_rate)

## Stochastic Gradient Descent

Problem Gradient Descent polega na tym, że wymusza na nas liczenie każdego przypadku modelu, tj. liczenie modelu dla każdej wartości Funkcji Straty. To przy dużej liczbie danych prowadzi do trylionów obliczeń, co jest dużą ilością nawet dla komputera jak na jedno dopasowanie modelu.

Tutaj z pomocą przychodzi Stochastic Gradient Descent. Jest to Gradient Descent tylko, że stochastyczny czyli przybliżony. Jest on stochastyczny, ponieważ dla każdej iteracji, nie liczymy SUMY KWADRATÓW RESZT dla wszystkich danych (dla parametru b i a) tylko dla jednego wylosowanego punktu lub określonej z góry ilości punktów jako hiperparametr w iteracji (mini-badge).Takie działanie znacznie przyspiesza działanie algorytmu, i powoduje niewielkie pogorszenie estymacji modelu, nadal będące bardzo dobrym dopasowaniem.

# Implementacja

Algorytm stochastycznego spadku do znalezienia minimum funkcji straty określonej wzorem: $L(w) = w^{2}-4w $. 

Pochodna tej funkcji to $\frac{dL}{dw}=2*w-4$ 

## Biblioteki

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()
```

## Generowanie danych

```python
w = np.arange(-10, 14, 0.1) # zwróci dane w zakresie od -10 do 14 (wyłączając) z częstością co 0.1
w[:20] # dane wyświetlane na osi x-ów
```

```python
loss = w**2 - 4*w # funkcja straty
loss[:20] # przedstawia wartości osi y
```

```python
point = -9
plt.plot(w, loss)
plt.plot(point, point**2 - 4 * point, 'ro')
plt.title('Loss Function')
```

## Algorytm gradientu

```python
def gd(df = lambda w: 2*w - 4, learning_rate = 0.01, w_0 = -9, max_iters = 10000, precision = 0.000001):
  '''
  deklaruję funkcję, która przyjmuje 3 argumenty:
  1. df czyli pochodną funkcji straty
  2. learning_rate domyślnie ustawiony na 0.01
  3. w_0 początkową wagę domyślnie ustawioną na -9, taką samą jak nasz point
  4. max_iter maxymalna ilość iteracji tak aby uniknąć pracy w nieskończoność
  5. precission odchylenie od optimum na jakim chcemy zakończyć pracę programu
  '''
  iters = 0 # licznik iteracji
  previous_step_size = 1 # kontrola wartości kroku kolejnego spadku

  # z tego powodu, że mamy ogranicznenie max_iters dla maximum iteracji oraz ograniczenie w postaci precission użyję pętli while
  while previous_step_size > precision and iters < max_iters:
    w_prev = w_0 # w_prev to poprzednia wartość wagi, zaczynamy od wagi w_0
    w_0 = w_0 - learning_rate * df(w_prev) # aktualizacja wagii, pamiętając, że df zadeklarowaliśmy jako argument funkcji za pomocą wyrażenia lambda
    previous_step_size = abs(w_0 - w_prev) # wielkość o jaką zaktualizowaliśmy naszą wagę
    iters += 1 # oznaczenie iteracji
    print('Iter #{}: obecny punkt: {}'.format(iters, w_0)) # wyświetlenie informacji o każdej iteracji (nr iteracji + waga)
  
  print('\nMinimum globalne: {}'.format(w_0))
```

## Graficzna reprezentacja treningu

Zupdatetujmy naszą klasę o możliwość przedstawiania graficznego jak nasze kolejne wagi były updateowane

```python
def gd(df = lambda w: 2*w - 4, learning_rate = 0.01, w_0 = -9, max_iters = 10000, precision = 0.000001):

  iters = 0 
  previous_step_size = 1 

  weights = [] # lista przechowująca kolejne zaktualizowane wagi

  while previous_step_size > precision and iters < max_iters:
    w_prev = w_0
    w_0 = w_0 - learning_rate * df(w_prev)
    previous_step_size = abs(w_0 - w_prev) 
    iters += 1 
    weights.append(w_0) # rozszerzamy naszą listę weights o naszą zaktualizowaną w tej iteracji wagę
  
  return weights 
```

```python
weights = gd(learning_rate = 0.1)
weights[:20] # wyświetlenie 20 pierwszych wag
```

```python
plt.figure(figsize = (8,6))
plt.plot(w, loss)
for weight in weights:
  plt.plot(weight, weight**2 - 4*weight, 'ro')
```

Jak widzimy nasz punkt powoli zbiegał do optimum. Dla zobrazowania czym jest learning rate pokażę poniżej jak zachowuje się traning dla bardzo małego learning_rate oraz bardzo dużego learning_rate

## Wyjaśnienie istotności hiperparametru learning_rate

```python
weights = gd(learning_rate = 0.00001)
plt.figure(figsize = (8,6))
plt.plot(w, loss)
for weight in weights:
  plt.plot(weight, weight**2 - 4*weight, 'ro')
```

W tym przypadku egzekucja programu trwała 16 sekund. Jak widzimy trening nie spowodował uzyskania maximum a to przez to, że learning_rate jest dużo zaniski. Program potrzebował poprostu tyle iteracji, że przekroczyło to nasze maximum ustalone we wskaźniku max_iter. Zwiększmy zatem nieco wskaźnik learning_rate.

```python
plt.figure(figsize = (8,6))
weights = gd(learning_rate = 0.001)
plt.figure(figsize = (8,6))
plt.plot(w, loss)
for weight in weights:
  plt.plot(weight, weight**2 - 4*weight, 'ro')
```

Egzekucja programu trwała 9 sekund. Jak widać doszliśmy do minimum bardzo małymi kroczkami. Przy bardzo dużych zestawach danych może to znacząco spowolnić działanie programu.

Teraz zwiększmy learning_rate

```python
weights = gd(learning_rate = 0.9)
plt.figure(figsize = (8,6))
plt.plot(w, loss)
for weight in weights:
  plt.plot(weight, weight**2 - 4*weight, 'ro')
```

Jak widzimy program doszedł do minimum jednak przez zbyt duży learning_rate punkt przeskakiwał z jednej strony osi na drugą.
